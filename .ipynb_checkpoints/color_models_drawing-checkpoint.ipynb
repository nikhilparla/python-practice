```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('burano.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_rgb, cmap = 'gray')

fig,axs = plt.subplots(nrows = 1, ncols = 3, figsize = (30,30))
for i in range(0,3):
    ax = axs[i]
    ax.imshow(img_rgb[:,:,i], cmap = 'gray')
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)

fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20,20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

```python
img_copy = img_rgb.copy()

# draw reactangle
cv2.rectangle(img_copy, pt1 = (400,300), pt2 = (600,500), color=(255,0,0), thickness = 5)
plt.imshow(img_copy)
```

```python
# add text
cv2.putText(img_copy, text = "The Boat", org = (400,300), fontFace = cv2.FONT_HERSHEY_DUPLEX, fontScale = 2, 
           color = (0,255,0), thickness = 2, lineType = cv2.LINE_AA)
plt.imshow(img_copy)
```
